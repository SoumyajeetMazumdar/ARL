# Print First Row of CSV Files in B/B-csv
This notebook iterates through all CSV files in the `B/B-csv` folder (relative to where you launch it) and prints each filename along with its first row (x, y).

```python
import os
import csv

# Path to the B-csv folder (inside B)
folder_path = os.path.join(os.getcwd(), 'pl', 'pl-csv')

# Iterate through CSV files and print filename + first row
for filename in sorted(os.listdir(folder_path)):
    if filename.lower().endswith('.csv'):
        filepath = os.path.join(folder_path, filename)
        with open(filepath, newline='') as csvfile:
            reader = csv.reader(csvfile)
            first_row = next(reader, None)
        print(f"File: {filename} -> First row: {first_row}")
```

```python
import numpy as np


# 1. Set your absolute output folder here:
output_dir = "/home/soumyajeet/Developer/ARL/PLS/ARL2025/pl/pl-plots"
os.makedirs(output_dir, exist_ok=True)

# 2. Prepare the x‐axis once
x = np.linspace(0, 2 * np.pi, 400)

# 3. Loop over k=1…10, plot and save
for k in range(1, 11):
    y = np.sin(k * x)
    plt.figure()
    plt.plot(x, y)
    plt.title(f"sin({k}·x)")
    plt.xlabel("x")
    plt.ylabel(f"sin({k}x)")
    
    out_path = os.path.join(output_dir, f"sin_{k}x.png")
    plt.savefig(out_path)
    plt.close()

print(f"Saved 10 plots to {output_dir}")
```
